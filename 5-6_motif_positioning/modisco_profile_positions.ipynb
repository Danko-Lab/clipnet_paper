```python
def get_counts(arr, fill_values=range(-200, 200)):
    counts = {i: list(arr).count(i) for i in arr if i in fill_values}
    for i in fill_values:
        if i not in counts:
            counts[i] = 0
    return pd.DataFrame(pd.Series(counts).sort_index()).T
```
