```python
import numpy as np
import seaborn as sns
from matplotlib.colors import LogNorm, Normalize
from matplotlib.ticker import MaxNLocator
import matplotlib.pyplot as plt
```

```python
dfim_profile=np.load("/Users/adamhe/github/scratch/dfim/active_full_length_sp1_tss_windows_dfim_profile_mean.npz")["arr_0"]
dfim_quantity=np.load("/Users/adamhe/github/scratch/dfim/active_full_length_sp1_tss_windows_dfim_quantity_mean.npz")["arr_0"]
```

```python
dfim_quantity_ = np.abs(
    dfim_quantity#.clip(min=np.percentile(dfim_quantity, 0.01), max=np.percentile(dfim_quantity, 99.99))
)[250-100:250+100, 500-100:500+100].transpose()
dfim_profile_ = np.abs(
    dfim_profile#.clip(min=np.percentile(dfim_profile, 0.01), max=np.percentile(dfim_profile, 99.99))
)[250-100:250+100, 500-100:500+100].transpose()

for i in range(dfim_quantity_.shape[0]):
    dfim_quantity_[i, i] = 0
    dfim_profile_[i, i] = 0
```

```python
sns.heatmap(
    dfim_quantity_,
    cmap="rocket_r",
    xticklabels=10, 
    yticklabels=10
)
```

```python
sns.heatmap(
    dfim_profile_,
    cmap="rocket_r",
    square=True,
    xticklabels=10, 
    yticklabels=10,
)
plt.savefig("sp1_dfim_profile.pdf")
```
