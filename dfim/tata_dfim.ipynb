```python
import numpy as np
import seaborn as sns
from matplotlib.colors import LogNorm, Normalize
from matplotlib.ticker import MaxNLocator
```

```python
dfim_fwd=np.load("/Users/adamhe/github/scratch/dfim/tata_fwd_dfim_profile.npz")
dfim_fwd=np.array([dfim_fwd[f] for f in dfim_fwd.files])
dfim_rev=np.load("/Users/adamhe/github/scratch/dfim/tata_rev_dfim_profile.npz")
dfim_rev=np.array([dfim_rev[f] for f in dfim_rev.files])
```

```python
dfim=np.concatenate([dfim_fwd, dfim_rev])
```

```python
sns.heatmap(
    dfim.mean(axis=0),
    cmap="rocket_r",
    xticklabels=10, 
    yticklabels=10
)
```

```python
sns.heatmap(
    np.log(dfim.mean(axis=0)),
    cmap="rocket_r",
    xticklabels=10, 
    yticklabels=10
)
```
