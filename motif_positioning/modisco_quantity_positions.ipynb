```python
"""
This notebook plots the positions of TF-MoDISco profile motifs relative to the TSS.
We mostly use this to identify the distribution of activator motifs around TSS (Fig. 5A).
"""
```

```python
import h5py
import seaborn as sns
import os
import matplotlib.pyplot as plt
```

```python
# PRINT COMMANDS TO DOWNLOAD AND EXTRACT DATA

# Set SCRATCH to where you want to download data to
SCRATCH = "/Users/adamhe/github/scratch"

URL = "https://zenodo.org/records/10597358/files"
TAR = "tfmodisco_results.tar.gz"
print(f"wget {URL}/{TAR} -P {SCRATCH}")
print(f"tar -xvzf {SCRATCH}/{TAR}")
```

```python
# Load data

hf = h5py.File(os.path.join(SCRATCH, "tfmodisco_results/mean_across_folds_all_modisco_positions.h5"))
pos = hf["quantity_modisco_positions"]["pos"]
neg = hf["quantity_modisco_positions"]["neg"]

# Set xticks
xticks = [-200, -100, -50, 0, 50, 100, 200]
```

```python
hf["quantity_modisco_positions"]["pos"]["0"][:]
```

```python
plt.rcParams["figure.figsize"] = 5,1
fig, (ax) = plt.subplots(nrows = 1, sharex = True)

# 2000
i = 0
sns.kdeplot(pos[str(i)][:], bw_adjust=0.5)
ax.set_xlim(-250, 250)
ax.set_xticks(xticks)
#ax.set_xticklabels([])
#ax.set_yticklabels([])
plt.savefig("img_quantity_modisco_position/ETS_kde.pdf", bbox_inches="tight")
```
