```python
import numpy as np
import os
import matplotlib.pyplot as plt
from utils import plot_weights
```

```python
# PRINT COMMANDS TO DOWNLOAD AND EXTRACT DATA

# Set SCRATCH to where you want to download data to
SCRATCH = "/Users/adamhe/github/scratch"

URL = "https://zenodo.org/records/10597358/files"
TAR = "example_tracks_and_deepshap.tar.gz"
print(f"wget {URL}/{TAR} -P {SCRATCH}")
print(f"tar -cvzf {SCRATCH}/{TAR}")
```

```python
# Load data

# SHAP scores
shap = np.load(
    os.path.join(SCRATCH, "example_tracks_and_deepshap/folds_qtl_examples_quantity.npz")
)
mean_shap = np.array([shap[fold] for fold in shap.files]).mean(axis=0)
```

```python
plot_weights(mean_shap[0, :, 250:750], neg_height=-0.4, pos_height=1.2, subticks_frequency=50)
plt.savefig("img/ensemble_deepshap_quantity_rs185220G.pdf")
```
