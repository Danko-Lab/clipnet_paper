```python
"""
This notebook plots DeepSHAP profile tracks for the DPR examples (Fig. 6C-D).
"""
```

```python
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
from utils import plot_weights
```

```python
# PRINT COMMANDS TO DOWNLOAD AND EXTRACT DATA

# Set SCRATCH to where you want to download data to
SCRATCH = "/Users/adamhe/github/scratch"

URL = "https://zenodo.org/records/10597358/files"
TAR = "example_tracks_and_deepshap.tar.gz"
print(f"wget {URL}/{TAR} -P {SCRATCH}")
print(f"tar -cvzf {SCRATCH}/{TAR}")
```

```python
# Load data

# SHAP scores
ensemble_shap = np.load(
    os.path.join(SCRATCH, "example_tracks_and_deepshap/mean_across_folds_dpr_profile.npz")
)["arr_0"]

# DPR scores (From the HARPE SVR-H model)
dpr_scores = pd.read_csv(
    os.path.join(SCRATCH, "example_tracks_and_deepshap/active_dpr_svrh.bed.gz"), sep="\t", header=None
)
dpr_desc = dpr_scores.loc[:,10].sort_values(ascending=False).index
```

```python
plot_weights(
    ensemble_shap[dpr_desc[1], :, 250:750],
    subticks_frequency=50, neg_height=-1e-4, pos_height=8e-4
)
plt.savefig("img/ensemble_deepshap_profile_irf1_prom.pdf")
```
